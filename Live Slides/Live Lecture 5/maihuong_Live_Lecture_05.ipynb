Chọn tất cả các dòng (slicing): cho phép lấy cả dòng chứa điểm chọn
- `loc`: location `df.loc[row_label,col_label]`, `series.loc[row_label]`
- `iloc`: index location `df.iloc[row_index,col_index]`, `series.iloc[row_index]`

Chọn 1 cell: chỉ cho lấy đúng điểm chọn

### 2.4 Masking

- df_name.loc[conditions]
- df_name[conditions]
- df_name.query(conditions) 

````python
tips_df.loc[tips_df['sex'] == 'Female'].head(5)
````

````python
tips_df.query("sex == 'Female'").head(5)
````

**Lọc ra thông tin khách hàng nữ có hút thuốc**

````python
condition = (tips_df['sex'] == 'Female') & (tips_df['smoker'] == 'Yes')

tips_df.loc[condition].head(5)
````

````python
tips_df.query("sex == 'Female' & smoker == 'Yes'").head(5)
````

**Lọc tất cả bản ghi chứa thông tin khách hàng là nam mà có tổng hóa đơn > 10$**


````python
tips_df.query("sex == 'Male' & total_bill > 10")
````

**Lọc tất cả bản ghi chứa thông tin khách hàng là nam không hút thuốc à dùng suất ăn cho 3 người trở lên**

````python
tips_df[(tips_df['sex'] == 'Male') & (tips_df['smoker'] == 'Yes') & (tips_df['size'] >= 3)]
````

**Lọc tất cả các bản ghi chứa thông tin khách hàng là nữ, ăn tối ở nhà hàng vào cuối tuần (t7,CN) và tip cho nhân viên > 5$**

````python
tips_df.query("sex =='Female' & time == 'Dinner' & day == ('Sat','Sun') & tip >= 5")
````

### 2.5 Operations

#### 2.5.1 Groupby

![](https://blog.dask.org/images/split-apply-combine.png)

````python
tips_df.head(3)  #Lấy đuôi: tail(), lấy random: sample()
````

Tính trung bình số tiền (chưa tip) mỗi giới tính khi ăn tại nhà hàng 

````python
tips_df.groupby('sex').total_bill.mean()
````

Tính trung bình số tiền (có tip) mỗi giới tính khi ăn tại nhà hàng

````python
tips_df['total_bill_tip'] = tips_df['total_bill'] + tips_df['tip'] 
tips_df.head(5)
````

````python
tips_df.groupby('sex').total_bill_tip.mean().to_frame()
````

Số tiền lớn nhất, số tiền nhỏ nhất và trung bình theo hóa đơn và theo số tiền tip theo giới tính

````python
tips_df.groupby('sex').agg({'total_bill': ['max', 'min', 'mean'], 'tip':['max', 'min', 'mean']})
````

````python
tips_df.groupby('sex').agg(mean_total_bill_tip = ('total_bill','mean'), mean_total_bill = ('total_bill','mean'))
````

Tính tổng doanh thu nhà hàng theo thời gian dùng bữa của khách hàng theo tips_df


````python
tips_df.groupby('time').agg(sum_bill = ('total_bill_tip', 'sum'))
````

Tính tổng doanh thu không tip của nhà hàng theo thời gian dùng bữa của khách hàng không hút thuốc vào CN

````python
tips_df.query("smoker == 'No' & day == 'Sun'").groupby('time').total_bill.sum().to_frame()
````

#### 2.5.2 Pivot

Tính trung bình số tiền mỗi giới tính khi ăn tại nhà hàng

````python
tips_df.pivot_table(values='total_bill_tip', index='sex', aggfunc='mean').plot.bar();
````

#### 2.5.3 Concatentation/Join/Merge

|Merge methods| SQL Join Name | Meaning| _merge|
|-|-|-|-|
|left| LEFT OUTER JOIN|Chỉ sử dụng keys bên trái|left_only|
|right| RIGHT OUTER JOIN|Chỉ sử dụng keys bên phải|right_only|
|outer|FULL OUTER JOIN|Sử dụng keys của cả 2 dataframes|both|
|inner|INNER JOIN|Chỉ sử dụng keys giao nhau của 2 dataframes|both|

![](https://lh3.googleusercontent.com/-n76c6dtr5sw/YBO5d-3PzGI/AAAAAAAAAh8/xiT6YIzePLEXArb8uU1f1vgg8JRXYeg8ACLcBGAsYHQ/image.png) 

````python
left = tips_df.head(6)[['total_bill', 'tip']].reset_index()
right = tips_df.loc[4:8][['tip', 'sex']].reset_index()
````

````python
left
````

````python
right
````

````python
left.merge(right, how='left', on=['index','tip'])
````

````python
left.merge(right, how='right', on=['index','tip'])
````
