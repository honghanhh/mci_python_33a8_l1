## 1. INTRODUCTION

### 1.1 What is Pandas? 

`Pandas = Panel Data`

Là thư viện mã nguồn mở phục vụ cho việc phân tích và xử lý dữ liệu; được thiết kế để làm việc dễ dàng và trực quan với dữ liệu có cấu trúc (dạng bảng, đa chiều, ...) và dữ liệu chuỗi thời gian

- Pandas phù hợp với nhiều loại dữ liệu khác nhau:
    - Dữ liệu dạng bảng, như trong bảng SQL hoặc bảng tính Excel.
    - Dữ liệu chuỗi thời gian theo thứ tự và không có thứ tự.
    - Dữ liệu ma trận tùy ý với nhãn hàng và cột...
- Dễ dàng thao tác, phân tích, xử lý và trực quan hoá dữ liệu:
    - Khả năng xử lý dữ liệu mất mát (NaN,...), nhiễu, ...
    - Khả năng thay đổi kích thước: chèn và xóa cột, dòng
    - Khả năng căn chỉnh dữ liệu tự động và rõ ràng
    - Khả năng phân tách, gộp nối, chuyển đổi, định hình các tập dữ liệu 1 cách linh hoạt giúp cho việc tổng hợp và phân tích dữ liệu nhanh gọn, dễ dàng hơn.
    - Khả năng tải và lưu trữ dữ liệu theo nhiều format khác nhau: .csv, .txt, .excel, .pkl, .hdfs5, ...
    - Khả năng xử lí dữ liệu dạng chuỗi.
    - Khả năng tích hợp tốt với các thư viện khác của Python: SciPy, Matplotlib, Seaborn, Plotly, Sklearn, ...

### 2.5 Operations

#### 2.5.1 Groupby

![](https://blog.dask.org/images/split-apply-combine.png)

- Syntax:`groupby()`


````python
tips_df.head(3) #tail(), sample()
````

Tinh trung binh so tien moi gioi tinh khi an tai nha nay

````python
tips_df.groupby('sex').total_bill.mean() #hàm mean: Hàm này sẽ trả về giá trị trung bình của tập dữ liệu được truyền vào
````

````python
tips_df['total_bill_tip'] = tips_df['total_bill'] + tips_df['tip']
tips_df.head(3)
````

````python
tips_df.groupby('sex').total_bill_tip.mean().to_frame() #thêm to_frame để đưa về dạng DataFrame
````

So tien lon nhat, nho nhat va trung binh hoa don va so tien tips theo gioi tinh

`Câu lệnh tổng hợp: .agg (#dictionary)`

````python
tips_df.groupby('sex').agg({'total_bill':['max','min','mean'], 'tip':['max','min','mean']}) #dictionary {key: [value]}
````

````python
tips_df.groupby('sex').agg(mean_total_bill_tip=('total_bill_tip','mean'), mean_total_bill=('total_bill','mean'))
````

`Có thể kết hợp groupby vs masking
df_name[conditions].group()`

1. Tinh tong doanh thu cua nha hang theo thoi gian dung bua cua khach theo `tips_df`
2. Tinh tong doanh thu k tinh tips cua nha hang theo gioi gian dung bua voi rieng cac khach hang khong hut thuoc vao ngay chu nhat

````python
tips_df.groupby ('time').total_bill_tip.sum().to_frame()
````

````python
tips_df[(tips_df['smoker'] == 'No') & (tips_df['day']== 'Sun')].groupby('time').total_bill.sum().to_frame()
````

#### 2.5.3 Concatenation/Join/Merge

|Merge methods| SQL Join Name | Meaning| _merge|
|-|-|-|-|
|left| LEFT OUTER JOIN|Chỉ sử dụng keys bên trái|left_only|
|right| RIGHT OUTER JOIN|Chỉ sử dụng keys bên phải|right_only|
|outer|FULL OUTER JOIN|Sử dụng keys của cả 2 dataframes|both|
|inner|INNER JOIN|Chỉ sử dụng keys giao nhau của 2 dataframes|both|

![](https://lh3.googleusercontent.com/-n76c6dtr5sw/YBO5d-3PzGI/AAAAAAAAAh8/xiT6YIzePLEXArb8uU1f1vgg8JRXYeg8ACLcBGAsYHQ/image.png) 

````python
left = tips_df.head(6)[['total_bill','tip']].reset_index()
right = tips_df.loc[4:8][['tip','sex']].reset_index()
````

````python
left
````

````python
right
````

````python
left.merge(right, how='left', on=['index','tip']) #truyền cụm key để tránh tách cột
````

````python
left.merge(right, how='right', on=['index','tip'])
````

````python
left.merge(right, how='inner', on=['index','tip'])
````

````python
left.merge(right, how='outer', on=['index','tip'])
````
