### 1. Thư viện Matplotlib 

Matplotlib (cụ thể trong bài học này ta quan tâm tới gói `Matplotlib.pyplot` hay được viết gọn là `plt`) là một gói Python được sử dụng cho đồ họa 2D.

Matplotlib có 2 objects chính:
- `Figure`: chứa nhiều axes
- `Axes`: nơi vẽ đồ thị, thực tế chính là cái đồ thị

Nói 1 cách dễ hiểu, `Figure` giống khung ảnh và 1 khung ảnh có thể không chứa hoặc chứa 1 đến nhiều bức ảnh, và `Axes` tương ứng với những bức ảnh trong cái khung ảnh.  

Các thành phần khác:
- `Title`: tên của đồ thị
- `X axis label`,`Y axis label`: 2 trục (ngang, đứng) của đồ thị
- `Major tick label`, `Minor tick label`: Tick chính chứa con số, tick nhỏ có thể hoặc không chứa con số
- `Legend`: Chú thích
- `Grid`: Lưới (dễ căn chỉnh)
- `Line plot`: Đồ thị đường
- `Scatter plot`: Đồ thị phân tán
- `Marker`: Những chấm tròn biểu diễn điểm dữ liệu trong đồ thị phân tán
- `Axis spine`: Đường ghi nhận ranh giới khu vực dữ liệu
- `Spine`: Đường kết nối các dấu ticks trên trục và ghi nhận ranh giới của khu vực dữ liệu.

**Tham khảo**: [Thư viện Matplotlib](https://matplotlib.org/stable/gallery/index.html)

### 2. Thư viện Seaborn

Seaborn là thư viện mở rộng được viết trên nền Matplotlib, từ đó kế thừa các chức năng biểu diễn dữ liệu từ Matplotlib. Vì vậy, những gì Seaborn làm được thì Matplotlib cũng làm được tuy nhiên không có chiều ngược lại, cũng có nghĩa là Seaborn không thể thay thế hoàn toàn được Matplotlib (Seaborn hiểu đơn giản là 1 đứa con được Matplotlib tạo ra).

**Tham khảo**: [Thư viện Seaborn](https://seaborn.pydata.org/examples/index.html)

Mỗi thư viện đều có những ưu điểm riêng: Nếu Seaborn sử dụng ít cú pháp hơn và có các chủ đề mặc định tuyệt đẹp thì Matplotlib dễ dàng tùy chỉnh hơn thông qua việc truy cập các lớp. Và trong bài viết này, chúng ta sẽ cùng nhau khám phá vẻ đẹp của 2 thư viện.

### 1. Đồ thị đường

#### 1.1. Đồ thị 1 đường

#### **Thực hành**: Vẽ đồ thị hình sin của 100 điểm trong khoảng từ -10 đến 10.

**Gợi ý**: Cho trước giá trị x, y. Sử dụng hàm `plot()` để vẽ đồ thị đường

```python
x = np.linspace(-10,10,100)
y = np.sin(x)
```

#### **Sử dụng Matplotlib**:

````python
x = np.linspace(-10,10,100)
y = np.sin(x)
plt.plot(x,y, linestyle='-.',color='green')
plt.title('Sin chart')
plt.xlabel('value2')
plt.ylabel('value1')
plt.grid()
plt.show()
````

Tham khảo 1 số cách thay đổi linestyle tại [đây](https://matplotlib.org/stable/_images/sphx_glr_linestyles_001.png).

#### **Sử dụng Seaborn**:

Lưu ý đầu vào của Seaborn là 1 DataDrame, còn đầu vào của đồ thị sử dụng Matplotlib là 1 Series.

````python
df = pd.DataFrame({'x':x,'sin':y})
df.head(10)
````

````python
sns.lineplot(data=df, x = df['x'],y = df['sin'])
plt.title('Sin chart')
plt.xlabel('value2')
plt.ylabel('value1')
plt.grid()
plt.show()
````

**Cách 1**

````python
plt.plot(x,sin, label = 'sin',color='red')
plt.plot(x,cos, label = 'cos',color='green')
plt.title('Sin/cos chart')
plt.xlabel('x value')
plt.ylabel('sin/cos value')
plt.grid()
plt.legend(loc='best')
plt.show()
````

**Cách 2**:

````python
ax = plt.axes()
ax.plot(x,sin,label='sin')
ax.plot(x,cos,label='cos')
ax.set(xlabel='x_label',ylabel='y_label',title='function plot')
plt.grid()
ax.legend()
plt.show()
````

````python
plt.figure(figsize=(15,5)) ##khung tranh rỗng -> subplot và subplots

## tạo ra 1 khung tranh cho cos
plt.subplot(1,2,1) #hàng nào, cột nào, axis nào
plt.plot(x, sin, c='red')
plt.title('Sin chart')
plt.xlabel('x value')
plt.ylabel('sin value')
plt.grid()

## tạo ra 1 khung tranh cho sin
plt.subplot(1,2,2)
plt.plot(x, cos,c='green')
plt.title('cos chart')
plt.xlabel('x value')
plt.ylabel('cos value')
plt.grid()
plt.show()
````

````python
fig, ax = plt.subplots(1,2,figsize=(15,5))
ax[0].plot(x,sin, color='yellow')
ax[0].set(xlabel='x_label',ylabel='y_label',title='sin plot')
ax[1].plot(x,cos, color='orange')
ax[1].set(xlabel='x_label',ylabel='y_label',title='cos plot')
plt.show()
````

### 5. Biểu đồ tròn (Pie chart)

#### **Thực hành**: Vẽ biểu đồ tròn thể hiện giới tính nam và nữ chiếm bao nhiêu phần trăm trong tổng hoá đơn.

````python
tips = sns.load_dataset('tips')
tips.head(5)
````

````python
sex = tips.groupby('sex').total_bill.sum().to_frame()
sex
````

````python
tips.groupby('sex').total_bill.count()
````

#### **Sử dụng Matplotlib**:

````python
plt.pie(sex, labels=sex.index, autopct='%.2f%%')
plt.show()
````

#### **Sử dụng Seaborn**:
Seaborn thì không hỗ trợ trực tiếp **pie chart**, ta có thể sử dụng **bar plot chart** thay thế.

````python
sex1 = sex.reset_index()
sex1
````

````python
sns.barplot(data=sex1,x='sex',y='total_bill')
````

Ta cũng có thể sử dụng barplot với thư viện Matplotlib và Pandas như dưới đây:

````python
plt.bar(data=sex1,x='sex', height='total_bill', color=['blue','orange'])
plt.show()
````

````python
#sử dụng pandas để visualize
sex1.plot(kind='bar')
plt.show()
````

**Bonus**: Ta có thể sử dụng `countplot()` để đếm số lượng, ví dụ:

````python
sns.countplot(data=tips,x='sex')
plt.show()
````

#### **BVN**: Vẽ biểu đồ tròn thể hiện giới tính nam và nữ chiếm bao nhiêu phần trăm trong tổng hoá đơn phía bên trái và biểu đồ tròn thể hiện tổng hoá đơn tiêu thụ theo ngày phía bên phải.

````python
left = tips.groupby('sex').total_bill.sum()
right = tips.groupby('day').total_bill.sum()
fig, ax = plt.subplots(1,2,figsize=(15,5))
ax[0].pie(left, labels=left.index, autopct='%.2f%%')
ax[1].pie(right, labels=right.index, autopct='%.2f%%')
ax[0].set_title('Sexuality/Total Bill')
ax[1].set_title('Day/Total Bill')
plt.show()
````

#### **Bài tập**: Vẽ biểu đồ tròn thể hiện phần trăm tổng hoá đơn tiêu thụ của nam theo ngày.

````python
tips.head(5)
#tính total_bill từng ngày
#giới tính nam
````

````python
df1 = tips[tips['sex']=='Male'].groupby('day').total_bill.sum()
df1
````

````python
plt.pie(df1,labels=df1.index,autopct='%.2f%%',explode=[0, 0.1, 0, 0])
plt.show()
````
