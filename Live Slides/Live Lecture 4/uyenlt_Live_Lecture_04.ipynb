#### 1.3 Một số hàm có sẵn hỗ trợ tạo mảng
Ngoài `np.array()`, ta có thể tạo 1 số array đặc thù với `np.zeros()`, `np.ones()`, `np.arange()`,`np.linspace()`

|Hàm|Ý nghĩa|
|-|-|
|`np.zeros()`|1 mảng chứa các giá trị 0|
|`np.ones()`|1 mảng chứa các giá trị 1|
|`np.full()`|1 mảng chứa các giá trị được gán theo `fill_value`|
|`np.arange()`|1 mảng chứa khoảng giá trị có khoảng cách đều nhau|
|`np.linspace`|1 mảng các giá trị có khoảng cách tuyến tính với nhau|

```python
#arange và linspace cần truyền parameter khác
#arange cần truyền số phần tử mong muốn -> tự tạo khoảng cách đều nhau
#linspace cần truyền số phần tử mong muốn -> tự tạo khoảng cách tuyến tính 
```

```python
np.zeros(shape=3, dtype = int) 
#1x3
```

```python
np.ones(shape=(3,3)) #(3,3) = rows,columns
```

```python
np.full(shape=(3,3), fill_value=100)
```

```python
np.arange(1,10,2, dtype=float) #only need to define distance of items in array
```

```python
np.linspace(1,10,5) #only need to define number of items in array
```

#### 1.4 Tạo mảng ngẫu nhiên

Trong NumPy, hàm `np.random` lấy ***ngẫu nhiên*** các giá trị trong 1 phân phối có ***quy luật***. 

```python
#pseudo-random: giả ngẫu nhiên -> do con người tạo ra 
```

```python
from datetime import datetime
datetime.now()
```

```python
x = datetime.now().microsecond
rand = 2*x+3 #x is always changing 
rand
```

Trong hàm `np.random`, chúng ta có 2 hàm con là `np.random.random()` và `np.random.normal()` tương ứng với 2 kiểu phân phối chúng ta đã quen thuộc khi học về xác suất thông kê, đó là:

- **Phân phối đồng nhất (Uniform Distribution)**
- **Phân phối chuẩn/ Phân phối hình chuông (Normal Distribution)** 

```python
#Phân phối đồng nhất (Uniform Distribution): tất cả các kết quả đều có xác suất xảy ra như nhau mà không có bias nào (vd: tung xúc sắc thì xác suất ra 1 mặt luôn bằng 1/6)
#Phân phối chuẩn/ Phân phối hình chuông (Normal Distribution): bell-curved distribution, giá trị phân phối được xác định bởi mean=0 (đại diện cho trung tâm phân bố dữ liệu, nếu lệch -> skewness, long-tail, outlier) và standard deviation=1 trong đó mean=median(giá trị phổ biến nhất)=mode(giá trị xuất hiện tần suất cao nhất)
```

```python
np.random.random([3,3])
```

```python
import matplotlib.pyplot as plt
plt.hist(np.random.random(10000000), bins=20) #100 for sample #bins cho độ phân biệt -> 20 columns
plt.show()
```

**Câu hỏi**: Vì sao hình trên không theo phân phối chuẩn?

```python
#có thể do sample size k đủ lớn, chưa thể hiện phân bổ đúng về data
#population: all data collected for the project
#sample: a set of data within the population but satisfied that it shares the same distribution as the population's
```

```python
np.random.normal([3,3])
```

```python
plt.plot(np.random.normal([3,9])) #100 for sample #bins cho độ phân biệt -> 20 columns
plt.show()
```
