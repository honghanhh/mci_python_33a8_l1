## 1.2. Shortcuts

- Run = Shift + Return
- Convert cell code to markdown = ESC + m
- Convert cell markdown to code = ESC + y
- Insert cell above: ESC + a
- Insert cell below: ESC + b
- Copy cell: ESC + c
- Cut cell: ESC + x
- Paste cell: ESC + v
- Delete cell: ESC + DD

- Định dạng văn bản:
    + Italic: Dùng dấu * hoặc _ ở đầu và cuối văn bản. E.g.: *italic* & *italic*
    + Bold: Dùng dấu ** hoặc __ ở đầu và cuối văn bản. E.g.: **bold** & __bold__
    + Bold & Italic: __*bold & italic*__ & **_bold & italic_**
