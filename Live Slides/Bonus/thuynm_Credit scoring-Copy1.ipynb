**Q12: Handle outliers in dataset**

- Quartile (tứ phân vị): Chia đồ thị thành 4 phần = (25%-Q1/ 50%-Q2/ 75%-Q3). interquartile = Vùng dữ liệu (Q3-Q1). Outlier là các điểm nằm ngoài interquartile (Q1-1.5.IQR,Q3 + 1.5.IQR). TH số biến nhỏ thì có thể chưa loại đi toàn bộ outlier mà set 1 ngưỡng nhất định: VD bản ghi nào có > 2 feature là outlier thì loại.
- Quantile
- Decile
- Percentile

```python
from collections import Counter
def detect_outliers(df, n, features): 
    outlier_indices =[]
    # lặp qua các features
    for col in features:
        # Tính 1st quartile
        Q1 = np.percentile(df[col],25)
        # Tính 3rd quartile
        Q3 = np.percentile(df[col],75)
        # Tính IQR
        IQR = Q3-Q1
        # Tính outlier 
        outlier_step = 1.5*IQR
        # Lấy index cột 
        outlier_list_col = df[(df[col]< Q1- outlier_step) | (df[col]> Q3 + outlier_step)].index
        outlier_indices.extend(outlier_list_col)
        
        
    # Chọn các bản ghi có tổng index chứa > n outlier
    outlier_indices = Counter(outlier_indices) # dictionary
    multiple_outlier = [k for k,v in outlier_indices.items() if v > n] # k,v: key/value trong dict
    
    return multiple_outlier
    
```
